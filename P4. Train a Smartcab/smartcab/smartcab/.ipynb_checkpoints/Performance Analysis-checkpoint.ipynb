```python
import pandas as pd
import numpy as np
from ggplot import *
%matplotlib inline
```

```python
df = pd.read_pickle('runHist.pkl')
```

```python
df[:21]
```

```python
temp = df.groupby(['trip']).agg({'deadline':'max','trip':'count'})
temp['deadline']+=1
ggplot(temp, aes(x='Trip Number',y='Steps to Complete')) + geom_line() + ggtitle('Number of Steps to Complete a Trip over Time')
```
